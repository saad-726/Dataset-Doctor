```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
#replace dataset path with yours!
df=pd.read_csv('/content/sample_data/california_housing_test.csv')
```

## Analyze Missing Values

### Subtask:
Calculate and display the percentage of missing values for each column in the DataFrame to identify columns with incomplete data.


**Reasoning**:
To calculate the percentage of missing values, I will use the `.isnull().sum()` method on the DataFrame `df` to count missing values per column, then divide by the total number of rows and multiply by 100.



```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("Percentage of missing values per column:")
print(missing_percentage)
```

## Analyze Target Variable Distribution

### Subtask:
Visualize the distribution of the 'median_house_value' using a histogram and a kernel density estimate (KDE) plot to understand its spread, skewness, and potential outliers. This helps in understanding the 'class imbalance' for a regression problem.


**Reasoning**:
To visualize the distribution of the 'median_house_value' column, I will use `seaborn.histplot` with `kde=True` to generate a histogram overlaid with a Kernel Density Estimate plot, and then customize the plot with appropriate titles and labels using `matplotlib.pyplot`.



```python
plt.figure(figsize=(10, 6))
sns.histplot(df['median_house_value'], kde=True, bins=50)
plt.title('Distribution of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()
```

## Identify Duplicate Rows

### Subtask:
Identify and report the number of duplicate rows in the DataFrame to ensure data integrity and avoid bias in analysis.

**Reasoning**:
To identify duplicate rows, I will use the `.duplicated()` method on the DataFrame `df` to create a boolean Series indicating duplicate rows, and then sum this Series to get the total count of duplicate rows. I will then print this count.



```python
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")
```

## Analyze Feature Correlations

### Subtask:
Compute and visualize the correlation matrix of all numerical features in the DataFrame `df` to understand relationships between variables and identify highly correlated features.

**Reasoning**:
To visualize the correlation matrix, I will first select all numerical columns from the DataFrame, calculate their correlations using the `.corr()` method, and then display the correlations as a heatmap using `seaborn.heatmap`. I will also ensure the plot is properly titled and formatted for readability.



```python
numerical_df = df.select_dtypes(include=np.number)
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
